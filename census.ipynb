```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample weather data
data = pd.read_csv("/content/sample_data/acs2017_census_tract_data.csv")

df = pd.DataFrame(data)

# Data Visualization
sns.lineplot(x='Women', y='Men', data=df)
plt.title("DISTRIBUTION")
plt.xlabel("Women")
plt.ylabel("Men")
plt.show()

sns.histplot(data=df, x='Women', kde=True)
plt.title("DISTRIBUTION")
plt.xlabel("Women")
plt.show()

```
