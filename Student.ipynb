```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample weather data
data = pd.read_csv("/content/sample_data/Private_data.csv")

df = pd.DataFrame(data)

# Data Visualization
sns.lineplot(x='RANK', y='PERCENTILE', data=df)
plt.title("PERFORMANCE")
plt.xlabel("RANK")
plt.ylabel("PERCENTILE")
plt.show()

sns.histplot(data=df, x='RANK', kde=True)
plt.title("PERFORMANCE")
plt.xlabel("RANK")
plt.show()

```
